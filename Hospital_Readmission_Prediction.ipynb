```python
# ===========================
# REUSABLE EDA FUNCTION
# ===========================

def analyze_variable(df, column, target="readmitted"):
    """
    Displays:
    1. Total Patients
    2. Readmitted <30
    3. Readmission Rate %
    """

    total = df[column].value_counts(dropna=False).sort_index()

    readmitted = (
        df[df[target] == "<30"][column]
        .value_counts(dropna=False)
        .sort_index()
    )

    summary = pd.DataFrame({
        "Patients": total,
        "Readmitted <30": readmitted
    }).fillna(0)

    summary["Readmitted <30"] = summary["Readmitted <30"].astype(int)

    summary["Readmission Rate %"] = (
        summary["Readmitted <30"] / summary["Patients"] * 100
    ).round(2)

    return summary
```

```python
# ==========================================================
# MASTER EDA FUNCTION
# ==========================================================

def analyze_variable(df, column, target="readmitted", min_patients=30):

    total = df[column].value_counts(dropna=False).sort_index()

    readmitted = (
        df[df[target] == "<30"][column]
        .value_counts(dropna=False)
        .sort_index()
    )

    summary = pd.DataFrame({
        "Patients": total,
        "Readmitted <30": readmitted
    }).fillna(0)

    summary["Readmitted <30"] = summary["Readmitted <30"].astype(int)

    summary["Readmission Rate %"] = (
        summary["Readmitted <30"] /
        summary["Patients"] * 100
    ).round(2)

    summary = summary[summary["Patients"] >= min_patients]

    print("="*90)
    print(f"VARIABLE : {column.upper()}")
    print("="*90)

    display(summary)

    plt.figure(figsize=(12,5))

    bars = plt.bar(
        summary.index.astype(str),
        summary["Readmission Rate %"]
    )

    plt.title(f"Readmission Rate by {column}", fontsize=14)
    plt.xlabel(column)
    plt.ylabel("Readmission Rate (%)")

    plt.xticks(rotation=45)

    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width()/2,
            height + 0.3,
            f"{height:.1f}%",
            ha="center",
            fontsize=8
        )

    plt.tight_layout()
    plt.show()

    return summary
```

```python
# FULL PREPROCESSING + BASELINE MODEL

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# 1. Copy clean data
data = df_processed.copy()

# 2. Create binary target
data["target"] = data["readmitted"].apply(lambda x: 1 if x == "<30" else 0)

# 3. Drop original target column
X = data.drop(columns=["readmitted", "target"])
y = data["target"]

# 4. Identify categorical and numerical columns
cat_cols = X.select_dtypes(include=["object"]).columns.tolist()
num_cols = X.select_dtypes(exclude=["object"]).columns.tolist()

print("Categorical columns:", len(cat_cols))
print("Numerical columns:", len(num_cols))

# 5. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

# 6. Preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
        ("num", "passthrough", num_cols)
    ]
)

# 7. Logistic Regression Model
log_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", LogisticRegression(max_iter=1000, class_weight="balanced"))
])

log_model.fit(X_train, y_train)

log_pred = log_model.predict(X_test)
log_prob = log_model.predict_proba(X_test)[:, 1]

print("\n==============================")
print("LOGISTIC REGRESSION RESULTS")
print("==============================")
print(confusion_matrix(y_test, log_pred))
print(classification_report(y_test, log_pred))
print("ROC-AUC:", roc_auc_score(y_test, log_prob))

# 8. Random Forest Model
rf_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        class_weight="balanced",
        n_jobs=-1
    ))
])

rf_model.fit(X_train, y_train)

rf_pred = rf_model.predict(X_test)
rf_prob = rf_model.predict_proba(X_test)[:, 1]

print("\n==============================")
print("RANDOM FOREST RESULTS")
print("==============================")
print(confusion_matrix(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
print("ROC-AUC:", roc_auc_score(y_test, rf_prob))
```

```python
# IMPROVED MODEL COMPARISON

from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

models = {
    "Logistic Regression": Pipeline([
        ("preprocessor", ColumnTransformer([
            ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
            ("num", StandardScaler(), num_cols)
        ])),
        ("model", LogisticRegression(max_iter=3000, solver="liblinear", class_weight="balanced"))
    ]),

    "Decision Tree": Pipeline([
        ("preprocessor", preprocessor),
        ("model", DecisionTreeClassifier(
            max_depth=8,
            min_samples_leaf=50,
            class_weight="balanced",
            random_state=42
        ))
    ]),

    "Random Forest": Pipeline([
        ("preprocessor", preprocessor),
        ("model", RandomForestClassifier(
            n_estimators=200,
            max_depth=12,
            min_samples_leaf=50,
            class_weight="balanced",
            random_state=42,
            n_jobs=-1
        ))
    ])
}

results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    prob = model.predict_proba(X_test)[:, 1]

    auc = roc_auc_score(y_test, prob)

    print("\n==============================")
    print(name)
    print("==============================")
    print(confusion_matrix(y_test, pred))
    print(classification_report(y_test, pred))
    print("ROC-AUC:", auc)

    results.append([name, auc])

results_df = pd.DataFrame(results, columns=["Model", "ROC-AUC"]).sort_values("ROC-AUC", ascending=False)
results_df
```

```python
# ==========================================================
# STEP 25: GROUP ICD-9 DIAGNOSIS CODES
# Objective:
# Convert diagnosis codes into disease categories.
# ==========================================================

def diagnosis_category(code):

    try:
        code = float(code)

        if 390 <= code <= 459 or code == 785:
            return "Circulatory"

        elif 460 <= code <= 519 or code == 786:
            return "Respiratory"

        elif 520 <= code <= 579 or code == 787:
            return "Digestive"

        elif 250 <= code < 251:
            return "Diabetes"

        elif 800 <= code <= 999:
            return "Injury"

        elif 710 <= code <= 739:
            return "Musculoskeletal"

        elif 580 <= code <= 629 or code == 788:
            return "Genitourinary"

        elif 140 <= code <= 239:
            return "Neoplasms"

        else:
            return "Other"

    except:
        return "Other"


df_processed["diag1_group"] = df_processed["diag_1"].apply(diagnosis_category)
df_processed["diag2_group"] = df_processed["diag_2"].apply(diagnosis_category)
df_processed["diag3_group"] = df_processed["diag_3"].apply(diagnosis_category)

print("Diagnosis grouping completed.")

print(df_processed[["diag_1","diag1_group"]].head())
```

```python
#STEP 38: Balance the Training Data Using SMOTE
#Objective

#Your training data is highly imbalanced:
#Class 0 (Not Readmitted) → ~88.8%
#Class 1 (Readmitted <30) → ~11.2%

##Important: Apply SMOTE only to the training set, never to the test set. Otherwise, you introduce data leakage and get overly optimistic results
```

```python
# ==========================================================
# STEP 38: APPLY SMOTE TO TRAINING DATA
# Objective:
# Balance the training data by oversampling the
# minority class using SMOTE.
# ==========================================================

# Install imbalanced-learn if needed
!pip -q install imbalanced-learn

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from imblearn.over_sampling import SMOTENC

# Identify categorical feature indices
categorical_indices = [X.columns.get_loc(col) for col in categorical_columns]

# One-Hot Encode is NOT done before SMOTE.
# Convert categorical columns to integer codes temporarily.

X_train_smote = X_train.copy()

for col in categorical_columns:
    X_train_smote[col] = X_train_smote[col].astype("category").cat.codes

smote = SMOTENC(
    categorical_features=categorical_indices,
    random_state=42
)

X_train_balanced, y_train_balanced = smote.fit_resample(
    X_train_smote,
    y_train
)

print("Original Training Shape:")
print(X_train.shape)

print("\nBalanced Training Shape:")
print(X_train_balanced.shape)

print("\nTarget Distribution After SMOTE:")
print(y_train_balanced.value_counts())
```

```python
# ==========================================================
# STEP 39: RANDOM FOREST AFTER SMOTE
# Objective:
# Train Random Forest on balanced data and
# evaluate its performance.
# ==========================================================

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.ensemble import RandomForestClassifier

# Convert categorical columns back to string
X_train_balanced = X_train_balanced.copy()

for col in categorical_columns:
    X_train_balanced[col] = X_train_balanced[col].astype(str)

X_test_model = X_test.copy()

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_columns),
        ("num", "passthrough", numerical_columns)
    ]
)

rf_model = Pipeline([
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(
        n_estimators=100,
        max_depth=15,
        random_state=42,
        n_jobs=-1
    ))
])

rf_model.fit(X_train_balanced, y_train_balanced)

rf_pred = rf_model.predict(X_test_model)
rf_prob = rf_model.predict_proba(X_test_model)[:,1]

print(confusion_matrix(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
print("ROC-AUC:", roc_auc_score(y_test, rf_prob))
```

```python
# THRESHOLD TUNING FOR RANDOM FOREST

from sklearn.metrics import precision_score, recall_score, f1_score

threshold_results = []

for threshold in [0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50]:
    pred_threshold = (rf_prob >= threshold).astype(int)

    threshold_results.append({
        "Threshold": threshold,
        "Precision": precision_score(y_test, pred_threshold),
        "Recall": recall_score(y_test, pred_threshold),
        "F1": f1_score(y_test, pred_threshold),
        "ROC-AUC": roc_auc_score(y_test, rf_prob)
    })

threshold_df = pd.DataFrame(threshold_results)
threshold_df
```

```python
# ==========================================================
# STEP 3: DROP HIGH MISSING COLUMNS
# Objective:
# Remove columns with excessive missing values.
# ==========================================================

df_ml.drop(
    columns=[
        "weight",
        "max_glu_serum",
        "A1Cresult"
    ],
    inplace=True
)

print(df_ml.shape)
```

```python
# ==========================================================
# STEP 4: HANDLE REMAINING MISSING VALUES
# Objective:
# Fill missing categorical values.
# ==========================================================

df_ml["race"] = df_ml["race"].fillna("Unknown")
df_ml["payer_code"] = df_ml["payer_code"].fillna("Unknown")
df_ml["medical_specialty"] = df_ml["medical_specialty"].fillna("Unknown")

for col in ["diag_1", "diag_2", "diag_3"]:
    df_ml[col] = df_ml[col].fillna("Unknown")

print(df_ml.isnull().sum().sum())
```

```python
# ==========================================================
# STEP 5: REMOVE IDENTIFIER COLUMNS
# Objective:
# Remove unique IDs.
# ==========================================================

df_ml.drop(
    columns=[
        "encounter_id",
        "patient_nbr"
    ],
    inplace=True
)

print(df_ml.shape)
```

```python
# ==========================================================
# STEP 6: CREATE BINARY TARGET
# Objective:
# Create binary readmission target.
# ==========================================================

df_ml["target"] = (
    df_ml["readmitted"] == "<30"
).astype(int)

print(df_ml["target"].value_counts())
```

```python
# ==========================================================
# STEP 2: LOAD DATASET
# Objective:
# Read the diabetic dataset into a DataFrame.
# ==========================================================

df = pd.read_csv("diabetic_data.csv")

print(df.shape)
```

```python
df_ml = df.copy()
```

```python
df_ml.replace("?", np.nan, inplace=True)

df_ml.drop(
    columns=["weight", "max_glu_serum", "A1Cresult"],
    inplace=True
)

df_ml[["race", "payer_code", "medical_specialty", "diag_1", "diag_2", "diag_3"]] = (
    df_ml[["race", "payer_code", "medical_specialty", "diag_1", "diag_2", "diag_3"]]
    .fillna("Unknown")
)

df_ml.drop(columns=["encounter_id", "patient_nbr"], inplace=True)

df_ml["target"] = (df_ml["readmitted"] == "<30").astype(int)

print(df_ml.shape)
print(df_ml["target"].value_counts())
```

```python
# CATBOOST MODEL - VERSION 2

!pip -q install catboost

from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

X = df_ml.drop(columns=["readmitted", "target"])
y = df_ml["target"]

cat_features = X.select_dtypes(include=["object"]).columns.tolist()

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

cat_model = CatBoostClassifier(
    iterations=500,
    learning_rate=0.05,
    depth=6,
    loss_function="Logloss",
    eval_metric="AUC",
    class_weights=[1, 8],
    random_seed=42,
    verbose=100
)

cat_model.fit(
    X_train,
    y_train,
    cat_features=cat_features,
    eval_set=(X_test, y_test)
)

cat_prob = cat_model.predict_proba(X_test)[:, 1]
cat_pred = (cat_prob >= 0.30).astype(int)

print(confusion_matrix(y_test, cat_pred))
print(classification_report(y_test, cat_pred))
print("ROC-AUC:", roc_auc_score(y_test, cat_prob))
```

```python
# ==========================================================
# THRESHOLD TUNING FOR CATBOOST
# Objective:
# Find the best probability threshold.
# ==========================================================

from sklearn.metrics import precision_score, recall_score, f1_score

results = []

for t in [0.20,0.25,0.30,0.35,0.40,0.45,0.50,0.55,0.60,0.65,0.70]:

    pred = (cat_prob >= t).astype(int)

    results.append([
        t,
        precision_score(y_test, pred),
        recall_score(y_test, pred),
        f1_score(y_test, pred)
    ])

threshold_results = pd.DataFrame(
    results,
    columns=["Threshold","Precision","Recall","F1"]
)

threshold_results
```

```python
# ==========================================================
# CATBOOST WITH AUTOMATIC CLASS WEIGHTS
# Objective:
# Compute class weights from the training data
# and train CatBoost using those weights.
# ==========================================================

from sklearn.utils.class_weight import compute_class_weight
from catboost import CatBoostClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
import numpy as np

# ----------------------------------------------------------
# Compute Class Weights
# ----------------------------------------------------------

classes = np.unique(y_train)

weights = compute_class_weight(
    class_weight="balanced",
    classes=classes,
    y=y_train
)

print("Computed Class Weights:", weights)

# ----------------------------------------------------------
# Train CatBoost
# ----------------------------------------------------------

cat_model = CatBoostClassifier(
    iterations=500,
    learning_rate=0.05,
    depth=6,
    loss_function="Logloss",
    eval_metric="AUC",
    class_weights=list(weights),
    random_seed=42,
    verbose=100
)

cat_model.fit(
    X_train,
    y_train,
    cat_features=cat_features,
    eval_set=(X_test, y_test)
)

# ----------------------------------------------------------
# Predictions
# ----------------------------------------------------------

cat_prob = cat_model.predict_proba(X_test)[:, 1]

# Change threshold if needed
threshold = 0.55

cat_pred = (cat_prob >= threshold).astype(int)

# ----------------------------------------------------------
# Evaluation
# ----------------------------------------------------------

print("\nConfusion Matrix")
print(confusion_matrix(y_test, cat_pred))

print("\nClassification Report")
print(classification_report(y_test, cat_pred))

print("\nROC-AUC:", roc_auc_score(y_test, cat_prob))
```
